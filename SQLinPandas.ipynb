```python
#Importing Pandas Library
import pandas as pd
```

```python
airportDf=pd.read_csv(r'C:\Users\736072\Downloads\airports.csv')
airportfreqDf=pd.read_csv(r'C:\Users\736072\Downloads\airport-frequencies.csv')
runwaysDf=pd.read_csv(r'C:\Users\736072\Downloads\runways.csv')
```

```python
#Basic SQL Commands to Understand Data
# SELECT COUNT(*) FROM AIRPORT
print (airportDf.count()) #Prints list of Not Null Values at Column level
print (airportDf.shape)
#print (airportfreqDf.count())
#print (runwaysDf.count())
```

```python
#Inspect few records from Dataframe
#SELECT * FROM AIRPORT LIMIT 10
print (airportDf.head(10)) #Prints list of Not Null Values at Column level
```

```python
# DISTINT COLUMN Values
#SELECT DISTINCT TYPE FROM AIRPORT
airportDf.iso_region.unique()
```

```python
#WHERE Condition
#SELECT * FROM AIRPORT WHERE ISO_REGION='MN-073'
airportDf[airportDf.iso_region=='MN-073']
```

```python
#Multiple WHERE Condition
#SELECT * FROM AIRPORT WHERE ISO_REGION='US-AK' AND TYPE='HELIPORT' AND ISO_COUNTRY='US'
airportDf[(airportDf.iso_region=='US-AK') & (airportDf.type=='heliport') & (airportDf.iso_country=='US')]
```

```python
#SELECTING Statements with Multiple Conditions
#SELECT ID, NAME FROM AIRPORT WHERE ISO_REGION ='US-AK' AND TYPE='HELIPORT'
airportDf[(airportDf.iso_region=='US-AK') & (airportDf.type=='heliport') & (airportDf.iso_country=='US')][['id', 'name']]
```

```python
#ORDER values based on Columns
#SELECT * FROM AIRPORT ORDER BY TYPE ASC LIMIT 5
airportDf.sort_values(by='type').head()

#SELECT ID, NAME, TYPE FROM AIRPORT ORDER BY TYPE ASC, ID ASC LIMIT 5
airportDf[['id', 'name', 'type']].sort_values(by=['type', 'id'], ascending=[True, True]).head()
```

```python
#IN
#SELECT COUNT(*) FROM AIRPORT WHERE TYPE IN ('HELIPORT', 'BALLLONPORT')
airportDf[airportDf.type.isin(['heliport', 'ballonport'])==True][['name']].count()
#SELECT DISTINCT MUNICIPALITY FROM AIRPORT WHERE TYPE IN ('HELIPORT', 'BALLLONPORT')
airportDf[airportDf.type.isin(['heliport', 'ballonport'])==True].loc[:, 'municipality'].unique()
```

```python
#NOT IN
#SELECT COUNT(*) FROM AIRPORT WHERE TYPE NOT IN ('HELIPORT', 'BALLLONPORT')
airportDf[airportDf.type.isin(['heliport', 'ballonport'])==False][['name']].count()
#SELECT DISTINCT MUNICIPALITY FROM AIRPORT WHERE TYPE NOT IN ('HELIPORT', 'BALLLONPORT')
airportDf[airportDf.type.isin(['heliport', 'ballonport'])==False].loc[:, 'municipality'].unique()
```

```python
#GROUP BY
#select iso_country, count(*) from airports group by iso_country
airportDf.groupby(by=['iso_country']).size()
#select iso_country, type, count(*) from airports group by iso_country, type order by iso_country, count(*) desc
airportDf.groupby(by=['iso_country', 'type']).size().to_frame('size').reset_index().sort_values(by=['iso_country', 'type'], ascending=True)
```

```python
#HAVING
#select type, count(*) from airports where iso_country = 'US' group by type having count(*) > 1000 order by count(*) desc
airportDf[airportDf.iso_country=='US'].groupby(by='type').filter(lambda g:len(g)>1000).groupby(by='type').size().sort_values(ascending=True)
```

```python
#JOINS
#join type: inner (default), left (corresponds to LEFT OUTER in SQL), right (RIGHT OUTER), or outer (FULL OUTER)
airportDf.merge(airportfreqDf,how='inner', left_on='id', right_on='airport_ref')[['id_x', 'name','description', 'type_y']]
```

```python
#left (corresponds to LEFT OUTER in SQL), right (RIGHT OUTER), or outer (FULL OUTER)
airportDf.merge(airportfreqDf,how='left', left_on='id', right_on='airport_ref')[['id_x', 'name','description', 'type_y']]
```

```python
#right (RIGHT OUTER), or outer (FULL OUTER)
airportDf.merge(airportfreqDf,how='right', left_on='id', right_on='airport_ref')[['id_x', 'name','description', 'type_y']]
```

```python
#JOINS with WHERE
airportDf.merge(airportfreqDf[airportfreqDf.description.isin(['TWR', 'ATIS'])==True],how='inner', left_on='id', right_on='airport_ref')[['id_x', 'name','description', 'type_y']]
```
